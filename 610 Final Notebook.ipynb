```python
def create_adjacency_matrix(G, nodes):
    """Creates the adjacency matrix from the graph."""
    matrix = np.zeros((len(nodes), len(nodes)))
    for i, node1 in enumerate(nodes):
        for j, node2 in enumerate(nodes):
            edge = G.get_edge_data(node1, node2)
            if edge is not None:
                matrix[i, j] = edge['weight']
    return matrix
#all interactions
all_interactions = (interactions_phantom_menace + interactions_attack_of_the_clones +
                    interactions_revenge_of_the_sith + interactions_a_new_hope +
                    interactions_empire_strikes_back + interactions_return_of_the_jedi +
                    interactions_the_force_awakens + interactions_the_last_jedi +
                    interactions_the_rise_of_skywalker)

#interactions
G = nx.Graph()
for character1, character2, strength in all_interactions:
    if G.has_edge(character1, character2):
        G[character1][character2]['weight'] += strength
    else:
        G.add_edge(character1, character2, weight=strength)

#adjacency matrix
nodes = list(G.nodes)
adj_matrix = create_adjacency_matrix(G, nodes)

#heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(adj_matrix, annot=True, fmt='g', xticklabels=nodes, yticklabels=nodes, cmap='YlGnBu')
plt.title("Figure 1: Adjacency Matrix of Frequency of Character Interactions")
plt.show()
```

```python
#specified nodes
def create_adjacency_matrix(G, nodes):
    """Creates the adjacency matrix from the graph for specified nodes."""
    matrix = np.zeros((len(nodes), len(nodes)))
    for i, node1 in enumerate(nodes):
        for j, node2 in enumerate(nodes):
            if node1 != node2 and G.has_edge(node1, node2):
                matrix[i, j] = G[node1][node2]['weight']
    return matrix
 


G = nx.Graph()
for character1, character2, strength in all_interactions:
    G.add_edge(character1, character2, weight=strength)

# top 10 characters
degree_sequence = sorted((d for n, d in G.degree(weight='weight')), reverse=True)
top_10_characters = sorted(G.degree(weight='weight'), key=lambda x: x[1], reverse=True)[:10]
top_10_characters = [character for character, _ in top_10_characters]

adj_matrix_top_10 = create_adjacency_matrix(G, top_10_characters)

plt.figure(figsize=(10, 10))
sns.heatmap(adj_matrix_top_10, annot=True, fmt='g', xticklabels=top_10_characters, yticklabels=top_10_characters, cmap='YlGnBu')
plt.title("Figure 1: Adjacency Matrix of Frequency of Character Interactions (Top 10 Characters)")
plt.show()
```
